```python
# Fraud Detection Analysis - Understanding Dataset and Fraud Rules

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('transactions.csv', parse_dates=['TX_DATETIME'])

# Preview data
print("\nHead of dataset:")
print(df.head())
print("\nDataset info:")
print(df.info())

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Fraud ratio
fraud_ratio = df['TX_FRAUD'].mean()
print(f"\nFraud ratio: {fraud_ratio:.4f}")

# Count of frauds
print("\nFraud counts:")
print(df['TX_FRAUD'].value_counts())

# Rule 1: High transaction amount frauds
print("\nRule 1: Transactions with amount > 220")
df['is_high_amount_fraud'] = df['TX_AMOUNT'] > 220
high_amount_frauds = df[df['is_high_amount_fraud'] == True]
print(f"Number of high-amount transactions: {high_amount_frauds.shape[0]}")

# Rule 2: Terminal-level fraud pattern
print("\nRule 2: Terminal-level fraud counts")
fraud_terminals = df[df['TX_FRAUD'] == 1]['TERMINAL_ID'].value_counts()
print(fraud_terminals.head())

# Rule 3: Customer-level inflated frauds
print("\nRule 3: Inflated customer transaction frauds")
customer_avg = df.groupby('CUSTOMER_ID')['TX_AMOUNT'].mean().rename('AVG_TX_AMOUNT')
df = df.merge(customer_avg, on='CUSTOMER_ID')
df['is_inflated'] = df['TX_AMOUNT'] > (df['AVG_TX_AMOUNT'] * 4.5)
inflated_frauds = df[df['is_inflated'] & (df['TX_FRAUD'] == 1)]
print(f"Number of suspected inflated frauds: {inflated_frauds.shape[0]}")
print(inflated_frauds.head())

# Visualization: Transaction amount distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TX_AMOUNT', hue='TX_FRAUD', bins=50, log_scale=(True, False))
plt.title('Transaction Amount Distribution by Fraud Label')
plt.xlabel('Transaction Amount')
plt.ylabel('Count')
# Only show legend if hue is used (which it is)
handles, labels = plt.gca().get_legend_handles_labels()
if labels:
    plt.legend(title='TX_FRAUD', labels=['Legit (0)', 'Fraud (1)'])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Ensure TX_DATETIME is datetime and sort values
df['TX_DATETIME'] = pd.to_datetime(df['TX_DATETIME'])
df = df.sort_values('TX_DATETIME')

# -----------------------------
# 🚨 High Transaction Amount Flag (Rule 1)
# -----------------------------
df['HIGH_AMOUNT_FLAG'] = df['TX_AMOUNT'] > 220

# -----------------------------
# 🏧 Terminal Fraud History (Rolling Count over 7 days)
# -----------------------------
def terminal_fraud_rolling(group):
    group = group.sort_values('TX_DATETIME')
    return group.set_index('TX_DATETIME')['TX_FRAUD'].rolling('7d').sum().fillna(0).reset_index(drop=True)

df['TERMINAL_FRAUD_COUNT_7D'] = (
    df.groupby('TERMINAL_ID')[['TX_DATETIME', 'TX_FRAUD']]
      .apply(terminal_fraud_rolling)
      .reset_index(drop=True)
)


# -----------------------------
# 👤 Customer Spending Patterns
# -----------------------------

# Average transaction amount per customer (up to current transaction)
def expanding_avg(group):
    return group['TX_AMOUNT'].expanding().mean().shift(1).fillna(0)

df['CUSTOMER_AVG_AMOUNT'] = (
    df.groupby('CUSTOMER_ID')['TX_AMOUNT']
      .apply(lambda x: x.expanding().mean().shift(1).fillna(0))
      .reset_index(drop=True)
)


# Deviation from customer average
df['CUSTOMER_TX_DEVIATION'] = df['TX_AMOUNT'] - df['CUSTOMER_AVG_AMOUNT']

# 7-day rolling average per customer
def customer_rolling_avg(group):
    group = group.sort_values('TX_DATETIME')
    return group.set_index('TX_DATETIME')['TX_AMOUNT'].rolling('7d').mean().shift(1).reset_index(drop=True)

df['CUSTOMER_TX_ROLLING_AVG_7D'] = (
    df.groupby('CUSTOMER_ID')[['TX_DATETIME', 'TX_AMOUNT']]
      .apply(customer_rolling_avg)
      .reset_index(drop=True)
)


# Fill missing with overall customer average if needed
df['CUSTOMER_TX_ROLLING_AVG_7D'] = df['CUSTOMER_TX_ROLLING_AVG_7D'].fillna(df['CUSTOMER_AVG_AMOUNT'])

# -----------------------------
# Final Check
# -----------------------------
print(df[['TX_DATETIME', 'TX_AMOUNT', 'HIGH_AMOUNT_FLAG',
          'TERMINAL_FRAUD_COUNT_7D', 'CUSTOMER_AVG_AMOUNT',
          'CUSTOMER_TX_DEVIATION', 'CUSTOMER_TX_ROLLING_AVG_7D']].head())

```

```python
# ✅ Step 3: Feature Engineering (Clean and Safe)

# Clean column names just in case
df.columns = df.columns.str.strip().str.upper()

# Ensure TX_DATETIME is datetime
df['TX_DATETIME'] = pd.to_datetime(df['TX_DATETIME'])
df = df.sort_values('TX_DATETIME')

# 🚨 High Transaction Amount Flag
df['HIGH_AMOUNT_FLAG'] = df['TX_AMOUNT'] > 220

# 🏧 Terminal Fraud Count - Rolling 7 Days
def terminal_fraud_rolling(group):
    group = group.sort_values('TX_DATETIME')
    return group.set_index('TX_DATETIME')['TX_FRAUD'].rolling('7d').sum().fillna(0).reset_index(drop=True)

df['TERMINAL_FRAUD_COUNT_7D'] = (
    df.groupby('TERMINAL_ID')[['TX_DATETIME', 'TX_FRAUD']]
      .apply(terminal_fraud_rolling)
      .reset_index(drop=True)
)


# 👤 Customer Average Amount (Expanding)
def expanding_avg(group):
    return group['TX_AMOUNT'].expanding().mean().shift(1).fillna(0)

df['CUSTOMER_AVG_AMOUNT'] = (
    df.groupby('CUSTOMER_ID')['TX_AMOUNT']
      .apply(lambda x: x.expanding().mean().shift(1).fillna(0))
      .reset_index(drop=True)
)


# Deviation from average
df['CUSTOMER_TX_DEVIATION'] = df['TX_AMOUNT'] - df['CUSTOMER_AVG_AMOUNT']

# Rolling average of last 7 days
def customer_rolling_avg(group):
    group = group.sort_values('TX_DATETIME')
    return group.set_index('TX_DATETIME')['TX_AMOUNT'].rolling('7d').mean().shift(1).reset_index(drop=True)

df['CUSTOMER_TX_ROLLING_AVG_7D'] = (
    df.groupby('CUSTOMER_ID')[['TX_DATETIME', 'TX_AMOUNT']]
      .apply(customer_rolling_avg)
      .reset_index(drop=True)
)

df['CUSTOMER_TX_ROLLING_AVG_7D'] = df['CUSTOMER_TX_ROLLING_AVG_7D'].fillna(df['CUSTOMER_AVG_AMOUNT'])

# ✅ Preview new features
df[['TX_DATETIME', 'TX_AMOUNT', 'HIGH_AMOUNT_FLAG', 'TERMINAL_FRAUD_COUNT_7D',
    'CUSTOMER_AVG_AMOUNT', 'CUSTOMER_TX_DEVIATION', 'CUSTOMER_TX_ROLLING_AVG_7D']].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# from imblearn.over_sampling import SMOTE
# from imblearn.under_sampling import RandomUnderSampler
# from collections import Counter
import numpy as np
```

```python
# -----------------------------
# Select Features & Target
# -----------------------------
features = [
    'TX_AMOUNT', 'HIGH_AMOUNT_FLAG', 'TERMINAL_FRAUD_COUNT_7D',
    'CUSTOMER_AVG_AMOUNT', 'CUSTOMER_TX_DEVIATION', 'CUSTOMER_TX_ROLLING_AVG_7D'
]
target = 'TX_FRAUD'

X = df[features].copy()
y = df[target]

# -----------------------------
# Convert boolean to integer (if needed)
# -----------------------------
X['HIGH_AMOUNT_FLAG'] = X['HIGH_AMOUNT_FLAG'].astype(int)

# -----------------------------
# Time-based Train-Test Split
# -----------------------------
split_date = df['TX_DATETIME'].quantile(0.80)  # Use 80% for training
X_train = X[df['TX_DATETIME'] < split_date]
X_test = X[df['TX_DATETIME'] >= split_date]
y_train = y[df['TX_DATETIME'] < split_date]
y_test = y[df['TX_DATETIME'] >= split_date]

print(f"Train size: {X_train.shape[0]}, Test size: {X_test.shape[0]}")
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# from sklearn.linear_model import LogisticRegression
# from sklearn.ensemble import RandomForestClassifier
# from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# # Define a function to train and evaluate a model
# def evaluate_model(name, model, X_train, y_train, X_test, y_test):
#     model.fit(X_train, y_train)
#     y_pred = model.predict(X_test)
#     y_proba = model.predict_proba(X_test)[:, 1]

#     print(f"\n📊 Model: {name}")
#     print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
#     print(f"Precision: {precision_score(y_test, y_pred):.4f}")
#     print(f"Recall:    {recall_score(y_test, y_pred):.4f}")
#     print(f"F1 Score:  {f1_score(y_test, y_pred):.4f}")
#     print(f"ROC AUC:   {roc_auc_score(y_test, y_proba):.4f}")

# # ---------------------------------
# # Define and evaluate models
# # ---------------------------------

# # 1. Logistic Regression (with class weights for imbalance)
# log_model = LogisticRegression(class_weight='balanced', max_iter=1000)
# evaluate_model("Logistic Regression", log_model, X_train_scaled, y_train, X_test_scaled, y_test)

# # 2. Random Forest
# rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
# evaluate_model("Random Forest", rf_model, X_train_scaled, y_train, X_test_scaled, y_test)
# rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
# evaluate_model("Random Forest", rf_model, X_train_scaled, y_train, X_test_scaled, y_test)


# plt.figure(figsize=(10, 6))
# for name, fpr, tpr, roc_auc in roc_data:
#     plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')
# plt.xlabel('False Positive Rate')
# plt.ylabel('True Positive Rate')
# plt.title('ROC Curve Comparison')
# plt.legend()
# plt.grid(True)
# plt.tight_layout()
# plt.show()


from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt

# Initialize storage for ROC curve data
roc_data = []

# Define a function to train, evaluate, and collect ROC data
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Evaluate performance
    print(f"\n📊 Model: {name}")
    print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred):.4f}")
    print(f"Recall:    {recall_score(y_test, y_pred):.4f}")
    print(f"F1 Score:  {f1_score(y_test, y_pred):.4f}")
    print(f"ROC AUC:   {roc_auc_score(y_test, y_proba):.4f}")

    # Store ROC data
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = roc_auc_score(y_test, y_proba)
    roc_data.append((name, fpr, tpr, roc_auc))

# Train and evaluate models
log_model = LogisticRegression(class_weight='balanced', max_iter=1000)
evaluate_model("Logistic Regression", log_model, X_train_scaled, y_train, X_test_scaled, y_test)

rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
evaluate_model("Random Forest", rf_model, X_train_scaled, y_train, X_test_scaled, y_test)

# Plot ROC Curves
plt.figure(figsize=(10, 6))
for name, fpr, tpr, roc_auc in roc_data:
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
